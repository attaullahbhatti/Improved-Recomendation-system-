```python
import csv
import re
x=set()
with open('selfmade.csv') as csv_file:
    csv_reader=csv.reader(csv_file,delimiter=",")
    for row in csv_reader:
        listofgenres=[]
        match = re.findall(r'\'([^\']*)\'', row[2])
        for item in match:
            x.add(item)
print(x)
for item in x:
    listofgenres.append(item)
print(listofgenres)
listofgenres.sort()
print(listofgenres)
```

```python
import csv
import re
finalbinary=[]
with open('selfmade.csv') as csv_file:
    csv_reader=csv.reader(csv_file,delimiter=",")
    for row in csv_reader:
        genresbinary=[]
        for standardgenres in listofgenres:
            if standardgenres in row[2]:
                genresbinary.append(1)
            else:
                genresbinary.append(0)
            
        finalbinary.append(genresbinary)
print(finalbinary)
                
                
                
            
```
