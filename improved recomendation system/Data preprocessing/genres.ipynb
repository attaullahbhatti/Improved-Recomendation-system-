```python
import csv
import re
allGenres=set()
with open('final.csv',encoding="UTF8") as csv_file:
    csv_reader=csv.reader(csv_file,delimiter=",")
    for row in csv_reader:
        match = re.findall(r'\'([^\']*)\'', row[0])
        for item in match:
            allGenres.add(item)
listofgenres=[]            
for item in allGenres:
    listofgenres.append(item)
#print(listofgenres)
listofgenres.sort()
#print(listofgenres)
```

```python
import csv
import re
import pandas as pd
finalbinary=[]
with open('final.csv',encoding="UTF8") as csv_file:
    csv_reader=csv.reader(csv_file,delimiter=",")
    next(csv_reader, None)
    for row in csv_reader:
        genresbinary=[]
        for standardgenres in listofgenres:
            if standardgenres in row[0]:
                genresbinary.append(1)
            else:
                genresbinary.append(0)
            
        finalbinary.append(genresbinary)
#print(finalbinary)
df=pd.read_csv('final.csv')
row,column=df.shape
for i in range(row):
    df['genres'][i]=finalbinary[i]
df.to_csv('final.csv',index=False)
                
                
                
            
```
